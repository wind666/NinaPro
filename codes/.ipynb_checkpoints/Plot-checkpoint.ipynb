```python
import numpy as np
from scipy.io import loadmat
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ninapro_DB1=loadmat("C:/Users/HP/Desktop/git_clone/NinaPro/unzip_data/"+'S' + str(1) + '_' + 'A1_E2.mat')
lab=(ninapro_DB1['restimulus'])
subject=(ninapro_DB1['subject'])
EMG=(ninapro_DB1['emg'])
label = [[row.flat[0] for row in line] for line in lab]
data = [[row.flat[0] for row in line] for line in EMG]
label=(np.ravel(label)).tolist()
np.shape(label),np.shape(data),np.shape(subject)
subject=subject[::2]
label=label[::2]
data=data[::2]
dataemg = {'subject': subject,
           'label':label,
           'data': data}
```

```python
def plot_channel(signal,time):
    fig = plt.figure(figsize=(20,20))
    for i in range(1,11):
        plt.subplot(5, 2, i)
        x=signal[i-1]
        plt.plot(time,x)
        #plt.plot(time,(dataemg['label'][:70000]), color='yellow', marker='.', linestyle='dashed',linewidth=0.2, markersize=0.5)
        plt.title("ch"+str(i))
    
    plt.show()
```

```python
s=np.array(dataemg['data'][0:70000]).T.tolist()
np.shape(s[1])
n_samples = 70000
time = np.linspace(0, 70, n_samples)
plot_channel(s,time)
```

```python
def plot_channel(signal,time):
    fig = plt.figure(figsize=(35,35))
    for i in range(1,11):
        fig, ax1 = plt.subplots()
        #plt.subplot(5, 2, i)

        x=signal[i-1]
        ax1.plot(time,x)
        ax1.set_ylabel('signal')
        ax2 = ax1.twinx()
        ax2.plot(time,(dataemg['label'][:70000]), color='yellow', marker='.', linestyle='dashed',linewidth=0.2, markersize=0.5)
        ax2.set_ylabel('label')
        plt.title("ch"+str(i))
    
    plt.show()
```

```python
s=np.array(dataemg['data'][0:70000]).T.tolist()
np.shape(s[1])
```
